```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
import os

cg = CoinGeckoAPI()

# Get top 10 coins and exclude stablecoins
top_coins = cg.get_coins_markets(vs_currency="usd", order="market_cap_desc", per_page=10, page=1)
coin_ids = [coin["id"] for coin in top_coins]
stablecoins = {"tether", "usd-coin", "staked-ether"}
coin_ids = [c for c in coin_ids if c not in stablecoins]
print("Filtered Top 10 coins:", coin_ids)

def fetch_coin(coin, days=365):
    data = cg.get_coin_market_chart_by_id(id=coin, vs_currency="usd", days=days)
    prices = pd.DataFrame(data["prices"], columns=["timestamp", "price"])
    prices["date"] = pd.to_datetime(prices["timestamp"], unit="ms")
    return prices.set_index("date")["price"]

# Always refresh cache completely when this cell is run
dfs = {}
for coin in coin_ids:
    dfs[coin] = fetch_coin(coin, days=365)

price_df = pd.concat(dfs, axis=1).dropna()
price_df.to_csv("crypto_prices.csv")
print("Data cached in crypto_prices.csv")
```

```python
import pandas as pd
import numpy as np
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt.risk_models import CovarianceShrinkage
from pypfopt.expected_returns import mean_historical_return
import matplotlib.pyplot as plt
from pypfopt import plotting

# Load cached data
price_df = pd.read_csv("crypto_prices.csv", index_col=0, parse_dates=True)
print("\nPrice Table (last 5 rows):")
print(price_df.tail())

# Compute returns & risk model
mu = mean_historical_return(price_df)
S = CovarianceShrinkage(price_df).ledoit_wolf()

# Optimization with constraints
ef = EfficientFrontier(mu, S)
btc, eth = "bitcoin", "ethereum"

# Constraint: BTC + ETH >= 70%
ef.add_constraint(lambda w: w[ef.tickers.index(btc)] + w[ef.tickers.index(eth)] >= 0.70)

# Constraint: other coins <= 10%
for coin in ef.tickers:
    if coin not in [btc, eth]:
        ef.add_constraint(lambda w, coin=coin: w[ef.tickers.index(coin)] <= 0.10)

# Portfolios
ef_min = ef.deepcopy()
w_min = ef_min.min_volatility()
cleaned_min = ef_min.clean_weights()

ef_sharpe = ef.deepcopy()
w_sharpe = ef_sharpe.max_sharpe()
cleaned_sharpe = ef_sharpe.clean_weights()

# Print results
print("\n--- Minimum Risk Portfolio ---")
for k,v in cleaned_min.items():
    if v > 0:
        print(f"{k}: {v:.2%}")
''
print("\n--- Maximum Sharpe Portfolio ---")
for k,v in cleaned_sharpe.items():
    if v > 0:
        print(f"{k}: {v:.2%}")

#Plot Efficient Frontier
plt.figure(figsize=(10, 7))

# Fresh EF for plotting
ef_plot = EfficientFrontier(mu, S)

# Re-apply constraints
ef_plot.add_constraint(
    lambda w: w[ef_plot.tickers.index(btc)] + w[ef_plot.tickers.index(eth)] >= 0.70
)

for coin in ef_plot.tickers:
    if coin not in [btc, eth]:
        ef_plot.add_constraint(
            lambda w, coin=coin: w[ef_plot.tickers.index(coin)] <= 0.10
        )

# Generate frontier points
n_points = 100
target_returns = np.linspace(mu.min(), mu.max(), n_points)

frontier_vols = []
frontier_rets = []
frontier_sharpes = []

for tr in target_returns:
    ef_tmp = EfficientFrontier(mu, S)

    # Re-apply constraints
    ef_tmp.add_constraint(
        lambda w: w[ef_tmp.tickers.index(btc)] + w[ef_tmp.tickers.index(eth)] >= 0.70
    )

    for coin in ef_tmp.tickers:
        if coin not in [btc, eth]:
            ef_tmp.add_constraint(
                lambda w, coin=coin: w[ef_tmp.tickers.index(coin)] <= 0.10
            )

    try:
        ef_tmp.efficient_return(tr)
        ret, vol, sharpe = ef_tmp.portfolio_performance()
        frontier_rets.append(ret)
        frontier_vols.append(vol)
        frontier_sharpes.append(sharpe)
    except:
        pass

plt.figure(figsize=(11, 7))

sc = plt.scatter(
    frontier_vols,
    frontier_rets,
    c=frontier_sharpes,
    cmap="viridis",
    s=25,
    edgecolors="black",
    alpha=0.9
)

cbar = plt.colorbar(sc)
cbar.set_label("Sharpe Ratio")

plt.plot(frontier_vols, frontier_rets, linewidth=2, color="black", alpha=0.6)


# Min volatility
r_min, v_min, _ = ef_min.portfolio_performance()
plt.scatter(v_min, r_min, marker="o", color="blue", s=120, label="Min Volatility")

# Max Sharpe
r_sh, v_sh, _ = ef_sharpe.portfolio_performance()
plt.scatter(v_sh, r_sh, marker="*", color="red", s=220, label="Max Sharpe")

risk_free_rate = 0.02  # annualized

cml_x = np.linspace(0, max(frontier_vols), 100)
cml_y = risk_free_rate + (r_sh - risk_free_rate) / v_sh * cml_x

plt.plot(
    cml_x,
    cml_y,
    linestyle="--",
    color="darkred",
    linewidth=2,
    label="Capital Market Line"
)

# Figure Plot

plt.xlabel("Volatility (Risk)", fontsize=12)
plt.ylabel("Expected Return", fontsize=12)
plt.title("Efficient Frontier with Sharpe Gradient and CML", fontsize=14)

plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

```
